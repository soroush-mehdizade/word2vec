# Importing necessary libraries

```python
import re
import nltk
import numpy as np
import pandas as pd
from gensim.models import word2vec
from nltk.corpus import reuters
from nltk.tokenize import word_tokenize
```

# Downloading stop words and puncuations by nltk

```python
nltk.download('stopwords')
nltk.download('punkt')
```

# Downloading the dataset 
the dataset is one of the nltk datasets used for nlp purposes which is a dataset of news articles labeled by topics.

```python
nltk.download('reuters')
```

```python
documents = [reuters.raw(fileid) for fileid in reuters.fileids()]
len(documents)
```

storing the stop words in stop_words variable

```python
stop_words = nltk.corpus.stopwords.words('english')
```

# Defining a function for preprocessing the documents
the function first lowers all the characters in the document ,then removes all non-alphabetic characters from the document 
after that it tokenizes the document into tokens and saves each individual word if the word isn't a stop word or has less
than 3 characters.

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    return tokens

sentences = [preprocess(doc) for doc in documents]
```

# Training the word2vec model using gensim library

```python
model = word2vec.Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=5,
    sg=1  # 1 = Skip-gram, 0 = CBOW
)
```

after training the model each word is shown as a vector of numbers as seen below: 

```python
model.wv['oil']
```

now we can use the created vectors to find similar words for example in the code below similar words to 'oil' have been shown 
along with the similarity rate that shows how similar each word is to 'oil'.

you can replace the oil with any arbitrary word to see its corresponding similar words.

```python
model.wv.most_similar("oil")
```

we can see the similarity rate for each pair of words using the .similarity() method.

you can replace the words with any arbitrary word to see the similarity of words.

```python
model.wv.similarity("bank", "money")
```
